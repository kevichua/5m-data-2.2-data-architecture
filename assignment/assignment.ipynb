Question: How would you create a new hash in Redis to store information about a person, with `john_doe` as the key and include the name (`John Doe`), age (`35`), and email address (`john@email.com`)?

```python
# Connect to Redis (adjust host/port if needed)
r = redis.Redis(host='localhost', port=6379, db=0)

# Create a new hash for john_doe
r.hset("john_doe", mapping={
    "name": "John Doe",
    "age": 35,
    "email": "john@email.com"
})

# Verify by retrieving the hash
person = r.hgetall("john_doe")
print(person)
```

Answer:

```python
# Answer Question 2 here
from google.cloud import storage

# Initialize the client
client = storage.Client()

# Reference the bucket
bucket = client.bucket("gcp-public-data-landsat")

# List the first 10 blobs
blobs = bucket.list_blobs(max_results=10)

# Print blob names and sizes
for blob in blobs:
    print(f"Name: {blob.name}, Size: {blob.size} bytes")
```
